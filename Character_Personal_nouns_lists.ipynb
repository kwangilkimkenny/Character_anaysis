personal nouns관련 단어들을 모두 모아서 txt파일로 만든다. 이것을 리스트로 만들어서 입력된(Essay)문장과 비교한다.

```python
import re
import nltk
from nltk.stem import PorterStemmer, LancasterStemmer #어간추출
```

```python
#txt 파일을 불러온다.
text_input=[]

file=open("data/personal_nouns.txt", 'r')

while (1):
    line=file.readline()

    try:escape=line.index('\n')
    except:escape=len(line)
    
    if line:
        text_input.append(line[0:escape])
    else:
        break
    
file.close()
```

```python
print(text_input[0:3])
```

```python
from nltk.corpus import stopwords
```

```python
def preprocessing(text_nouns, remove_stopwords=False):
        
    # 영어가 아닌 특수문자들을 공백(" ")으로 바꾸기
    #processed_text = re.sub("[^a-zA-Z]", " ", text_nouns)
 
    #텍스트에 포함되어 있는 특수 문자 제거
 
    processed_text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', text_nouns)
 

    # 대문자들을 소문자로 바꾸고 공백단위로 텍스트들 나눠서 리스트로 만든다.
    words =  processed_text.lower().split()

    if remove_stopwords: 
        # 불용어들을 제거
    
        #영어에 관련된 불용어 불러오기
        stops = set(stopwords.words("english"))
        # 불용어가 아닌 단어들로 이루어진 새로운 리스트 생성
        words = [w for w in words if not w in stops]
        # 단어 리스트를 ,를 넣어서 하나의 글로 합친다.
        clean_nouns_texts= ' '.join(words)

    else: # 불용어 제거하지 않을 때
        clean_nouns_texts = ' '.join(words)

    return clean_nouns_texts
```

```python
clean_data_ = []
for review in text_input:
    #list to str
    str_data = ' '.join(text_input)
    clean_data_.append(preprocessing(str_data, remove_stopwords=True))

# 전처리된 데이터 확인. 잘됨 !! ㅎ
clean_data_[0]
```
